```python
def text_list(c):
  txt_list=[]
  length=c.shape[0]
  for i in range(0,length):
    txt=' '.join([word for word in c.iloc[i]])
    txt_list.append(txt)
  return txt_list
```

```python
#tokenization of text
def Tokenize_text(txt_list):
  tokenize_text=[]
  for i in txt_list:

    tokenize_text+=(word_tokenize(i))
  return tokenize_text
```

```python
def Positive_score(tokenize_text):
  positive_score=0
  for i in tokenize_text:
    if(i.lower() in post):
      positive_score+=1
  return positive_score
print('postive score=', positive_score)
```

```python
def Negative_score(tokenize_text):
  negative_score=0
  for i in tokenize_text:
    if(i.lower() in neg):
      negative_score+=1
  return negative_score
print('negative score=', negative_score)
```

```python
#Polarity Score = (Positive Score – Negative Score)/ ((Positive Score + Negative Score) + 0.000001)
def polarity_Score(positive_score,negative_score):
  Polarity_Score=(positive_score-negative_score)/((positive_score+negative_score)+0.000001)
  return Polarity_Score
print('polarity_score=', Polarity_Score)
```

```python
#Subjectivity Score = (Positive Score + Negative Score)/ ((Total Words after cleaning) + 0.000001)
def subjectivity_score(positive_score,negative_score,tokenize_text):
  subjectivity_score=(positive_score+negative_score)/((len(tokenize_text))+ 0.000001)
  return subjectivity_score
print('subjectivity_score',subjectivity_score)
```

```python
def Avg_senetence_length(c):
  length=c.shape[0]
  avg_length=[]
  for i in range(0,length):
    avg_length.append(len(c['abc'].iloc[i]))
  avg_senetence_length=sum(avg_length)/len(avg_length)
  return avg_senetence_length
print('avg sentence length=', avg_senetence_length)
```

```python
def percentage_complex_words(tokenize_text):
  vowels=['a','e','i','o','u']
  count=0
  complex_Word_Count=0
  for i in tokenize_text:
    x=re.compile('[es|ed]$')
    if x.match(i.lower()):
      count+=0
    else:
      for j in i:
        if(j.lower() in vowels ):
          count+=1
    if(count>2):
      complex_Word_Count+=1
    count=0
  Percentage_of_Complex_words=complex_Word_Count/len(tokenize_text)
  return Percentage_of_Complex_words
print('percentag of complex words= ',percentage_complex_words)
```

```python
#Fog Index = 0.4 * (Average Sentence Length + Percentage of Complex words)
def fog_index(avg_senetence_length,Percentage_of_Complex_words):
  Fog_Index = 0.4 * (avg_senetence_length + Percentage_of_Complex_words)
  return Fog_Index
print('fog index= ',fog_index )
```

```python
def Avg_no_of_words_per_sentence(c):
  length=c.shape[0]
  avg_length=[]
  for i in range(0,length):
    a=[word.split( ) for word in c.iloc[i]]
    avg_length.append(len(a[0]))
    a=0
#avg
  avg_no_of_words_per_sentence=sum(avg_length)/length
  return avg_no_of_words_per_sentence
print("avg no of words per sentence= ",avg_no_of_words_per_sentence)
```

```python
def complex_words(tokenize_text):
  vowels=['a','e','i','o','u']
  import re
  count=0
  complex_Word_Count=0
  for i in tokenize_text:
    x=re.compile('[es|ed]$')
    if x.match(i.lower()):
      count+=0
    else:
      for j in i:
        if(j.lower() in vowels ):
          count+=1
    if(count>2):
      complex_Word_Count+=1
    count=0
  return complex_Word_Count
print('complex words count=',  complex_words)
```

```python
def Word_count(tokenize_text):
  word_count=len(tokenize_text)
  return word_count
print('word count= ', word_count)
```

```python
def Syllable_count(tokenize_text):
  vowels=['a','e','i','o','u']
  import re
  count=0
  for i in tokenize_text:
    x=re.compile('[es|ed]$')
    if x.match(i.lower()):
      count+=0
    else:
      for j in i:
        if(j.lower() in vowels ):
          count+=1
  syllable_count=count
  return syllable_count
print('syllable count= ', syllable_count)
```

```python
def Personal_pronouns(tokenize_text):
  pronouns=['i','we','my','ours','us' ]
  import re
  count=0
  for i in tokenize_text:
    if i.lower() in pronouns:
      count+=1
  personal_pronouns=count
  return personal_pronouns
print('personal pronouns= ',personal_pronouns )
```

```python
def Avg_word_length(tokenize_text):

  count=0
  for i in tokenize_text:
    for j in i:
      count+=1
  avg_word_length=count/len(tokenize_text)
  return avg_word_length
print('avg word= ', avg_word_length)
```

```python
def process_text(file_path):

  text = pd.read_csv(file_path, header=None)
  text = text.astype(str)
  split_text = splittext(text)
  split_text['abc']=split_text['abc'].apply(abcd)
  c=replace_sp(split_text)
  c['abc']=c['abc'].apply(text_process)
  txt_list=text_list(c)
  tokenize_text=Tokenize_text(txt_list)
  positive_score=Positive_score(tokenize_text)
  negative_score=Negative_score(tokenize_text)
  Polarity_Score=polarity_Score(positive_score,negative_score)
  subjectiivity_score=subjectivity_score(positive_score,negative_score,tokenize_text)
  avg_senetence_length=Avg_senetence_length(c)
  Percentage_of_Complex_words=percentage_complex_words(tokenize_text)
  Fog_Index=fog_index(avg_senetence_length,Percentage_of_Complex_words)
  avg_no_of_words_per_sentence=Avg_no_of_words_per_sentence(c)
  complex_Word_Count=complex_words(tokenize_text)
  word_count=Word_count(tokenize_text)
  syllable_count=Syllable_count(tokenize_text)
  personal_pronouns=Personal_pronouns(tokenize_text)
  avg_word_length=Avg_word_length(tokenize_text)
  data={'positive_score':positive_score,'negative_score':negative_score,'Polarity_Score':Polarity_Score,'subjectiivity_score':subjectiivity_score,'avg_senetence_length':avg_senetence_length,'Percentage_of_Complex_words':Percentage_of_Complex_words,'Fog_Index':Fog_Index,'avg_no_of_words_per_sentence':avg_no_of_words_per_sentence,'complex_Word_Count':complex_Word_Count,'word_count':word_count,'syllable_count':syllable_count,'personal_pronouns':personal_pronouns,'avg_word_length':avg_word_length}
  return pd.DataFrame(data,index=[0])

```

```python
text_files_dir = "/content/"
text_files = [os.path.join(text_files_dir, f) for f in os.listdir(text_files_dir) if f.endswith('.txt')]
```

```python
output = pd.DataFrame()
for file in text_files:
    result = process_text(file)
    output = pd.concat([output, result], ignore_index=True)
```
